# **Black Friday dataset**
* Objective: Cleaning and preparing data for model training

**Problem Statement:**
A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing dataset
df_train = pd.read_csv('train.csv')
#df_test = pd.read_csv('test.csv')
```

```python
df_train.columns
```

```python
#importing test data
df_test = pd.read_csv('test.csv')
df_test.head(2)
```

```python
df_test.columns
```

```python
#merging the two dataset
df = pd.concat([df_train, df_test], ignore_index=True) #we can also use merge function here
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#here user id is of no use for our model , it is unique for everyone, so dropping it
df.drop(['User_ID'], axis=1, inplace=True)
```

```python
df.head(15)
```

* Now we are trying to resolve the categorical features rightnow.i.e

1.   Gender
2.   Age(bcz given in range)
3.   City  category , etc

* So we'll try to change this categorical data into umber, just like martial status which is already given in 0/1



```python
#resolving gender : either it would be male or female
pd.get_dummies(df['Gender'])
```

* 0 for female is true
* 1 for male is true

```python
#instead of creating the new data frame we are manipulating data for gender column based on above values
df['Gender'] = df['Gender'].map({'F':0, 'M':1 })
df.head()
```

* Now the gender is either 0 or 1 rather than M/F

## Handling categorical feature

```python
df['Age'].unique()
```

* In order to change this categorical data into numbers, we use encoding:

```python
# Technique 1: fir encoding.

#pd.get_dummies(df['Age'],drop_first= True)
df['Age']=df['Age'].map({'0-17':1, '18-25':2, '26-35':3, '36-45':4, '46-50':5, '51-55':6, '55+':7})
```

```python
 #Technique 2 : for encoding
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df['Age']=le.fit_transform(df['Age'])
df['Age'].unique()
```
