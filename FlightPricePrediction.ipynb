```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

* Reading the dataset

```python
#reading train data in train_df
train_df = pd.read_excel('fptrain.xlsx')
#reading test data in test_df
test_df = pd.read_excel('fptest.xlsx')
```

```python
#now appending both data(appends rows of the 2 data)
final_df = pd.concat([train_df,test_df])
final_df.head()
```

## Handling Categorical data

* Now almost all data is of object which may be string or categorical
* as first feature is str and can't be converted to int/float nor any categorical data so leave it
* But our 'Date_of_Journey' is str however it can be change to int as dd mm yr (in 3 col)

**Handling F-1 : date of journey**

```python
#Creating 3 col as day, name , yr and splitting the data by '/' , and accessing each split part through index
final_df['Day'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(2)
```

* Now as we can see we have got 3 new col at the end and our col : date of journey is now useless so we can drop it

```python
# dropping doj
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(2)
```

```python
final_df.info()
```

* now the problem is however we have got day,month, year but they are still obj , so typecast it to int

```python
#typecasting
final_df['Day'] = final_df['Day'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

* Now these 3 col are int.


--------------------------------------------------------------------------------------------------------------------------------------------------------
**Handling F-2 : Arrival time**
* Now after it we'll go for any feature , i'm going for arrival time

```python
final_df.head(3)
```

* Now as we see the arrival time in some row just have time while some also have date , but we have date already so we can split them by ' ' and then just store the time in 'Arrival_time' ( as it is req only)

```python
final_df['Arrival_Time']= final_df['Arrival_Time'].str.split(' ').str[0] #at 0 index we have time at 1 date so taking 0 only
final_df.head()
```

* Now in arrival time we just have time but it is still str , which we have to convert to int so we'll break it into 2 col one for hr , one for min

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0].astype(int)
final_df['Arrival_Min']= final_df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
final_df.head(2)
```

* Now we have arr-hr and arr-min so we can drop arr-time

```python
final_df.drop('Arrival_Time',axis = 1, inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

* Now our arrival-time is in int, Let's hop to next feature, let's do it for Dept_time

**Handling F-3 : Dept_Time**

```python
final_df['Dept_Hr']= final_df['Dep_Time'].str.split(':').str[0].astype(int)
final_df['Dept_Min']= final_df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```
