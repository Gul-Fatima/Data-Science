## Finding relationships between columns:

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

* now this groupby is grouping the values for these 3 columns, .size gives the no of occurence i.e for 0.0= white we have 2148 records
* then we are using rename function to rename our generated column to rating count, otherwise it would name it 0(automatically)

```python
# saving the value in variable:
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

### Observations:
* rating from 4.5 to 4.9 is excellent food
* rating from to 4.0 to 4.4 is very good
* rating from 3.5 to 3.9 is good
* ratimg from 2.5 to 2.4 is average
* rating from 1.8 to 2.4 is poor
* usually people won't rate

```python
#visualizing the data
plt.matplotlib.rcParams['figure.figsize'] = (12,6) #resize the image
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings,hue='Rating color',palette=['pink','red','orange','yellow','seagreen','green'])
```

### Observations:


1.  Mostly not rated
2.  Max rating b/w 2.5 to 3.4(so if there are null values so we could give the values b/w this range as an average)


```python
#Count plot
sns.countplot(x='Rating color', data=ratings, palette=['pink','red','orange','yellow','seagreen','green'])
```

* above plot is not based on 'Rating count' instead it is about the occurences i.e the white occured one time so smallest
